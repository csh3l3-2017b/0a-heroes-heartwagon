### Heroes Membutuhkan Anda

Setiap <i>heroes</i> akan mengisi formulir biodata <i>hero</i>, dalam file teks bernama <code>test.csv</code> dengan struktur CSV

```python
#import panda untuk mempermudah membaca file .csv
import pandas as pd
```

```python
#membaca file test.csv dan menampilkan datanya
test_df = pd.read_csv("Downloads/test.csv")
test_df
```

Misi Anda adalah menentukan, apa peran dari masing-masing hero yang membutuhkan bantuan dari program Anda, dengan aturan sebagai berikut secara berurutan (hero mendapatkan peran yang <b>paling awal</b> memenuhi syarat):
1. Heroes yang memiliki <b>movementSpeed</b> >= 270, akan mendapat peran sebagai <b>MARKSMAN</b> (huruf kapital semua)
2. Heroes dengan <b>mana</b> >= 500, akan mendapat peran sebagai <b>MAGE</b> (huruf kapital semua)
3. Heroes dengan <b>armor</b> >= 25, akan mendapat peran sebagai <b>TANK</b> (huruf kapital semua)
4. Heroes dengan <b>physicalAttack</b> >= 121, akan mendapat peran sebagai <b>ASSASSIN</b> (huruf kapital semua)
5. Heroes dengan <b>hp</b> >= 2580, akan mendapat peran sebagai <b>FIGHTER</b> (huruf kapital semua)
6. Heroes lainnya akan mendapat peran sebagai <b>SUPPORT</b> (huruf kapital semua)


```python
#variabel untuk menyimpan peran hero
kelas = []
```

Selanjutnya buat fungsi untuk mengklasifikasikan hero berdasarkan status hero sesuai aturan diatas

```python
for data in range(len(test_df)) :
    if test_df["movementSpeed"][data] >= 270:
        kelas.append("MARKSMAN")
    elif test_df["mana"][data] >= 500:
        kelas.append("MAGE")
    elif test_df["armor"][data] >= 25:
        kelas.append("TANK")
    elif test_df["physicalAttack"][data] >= 121:
        kelas.append("ASSASSIN")
    elif test_df["hp"][data] >= 2580:
        kelas.append("FIGHTER")
    else: kelas.append("SUPPORT")
```

Langkah selanjutnya adalah gabungkan nama-nama hero dan perannya masing masing kedalam sebuah data frame baru

```python
prediksi = pd.DataFrame({
    "name" : test_df["name"],
    "role" : kelas
})
```

Kemudian hasilnya dikeluarkan ke file prediction.csv

```python
prediksi.to_csv("prediction.csv", index = False)
```
